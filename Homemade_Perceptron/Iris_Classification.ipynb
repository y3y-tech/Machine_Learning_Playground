## Data Filtering 

```python
import os
import pandas as pd
import numpy as np

s = 'https://archive.ics.uci.edu/ml/'\
'machine-learning-databases/iris/iris.data'
    
df = pd.read_csv(s,header=None,encoding='utf-8')

df.tail()
```

```python
df.describe()
```

```python
df_filtered = df[df[4] != 'Iris-virginica']
```

## Plotting

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.scatter(df_filtered[0],df_filtered[3],c=df_filtered[4].map({'Iris-setosa':0,'Iris-versicolor':1}),cmap='bwr')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Width')
plt.title('Iris Dataset - Sepal Length vs Petal Width')
plt.show()
```

```python
from Naive_Perceptron import Naive_Perceptron

model = Naive_Perceptron()

X = df_filtered[[0,3]].to_numpy()
Y = np.where(df_filtered[4].to_numpy() == "Iris-setosa", 1, 0)

model.fit(X,Y)
model.predict(X)

params = model.getParams()
```
